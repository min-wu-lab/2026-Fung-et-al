```python
def WT_panorama(dist, thickness):
    
    ### load point cloud data =================================================================================
#     fn_WT = 'F:/Acquisition_Yujin_3/Collaboration/Wu_Lab_Collaboration/20241114_RBL_STLC/WT/20241023_seq1.hdf'
    points = pipeline(fn_WT)
#     x0, y0 = 631, 772
#     x1, y1 = 189, 1334
#     ps = 5e4/1163
    xc = (x0wt-x1)*ps
    yc = (y1-y0wt)*ps
    points_x = points['x'] - xc
    points_y = points['y'] - yc
    points_z = points['z']
    mid_z = (points_z.min() + points_z.max())/2
    zmin = mid_z - 2e3
    zmax = mid_z + 2e3
    maskz = (points_z>=zmin) & (points_z<=zmax)
    points_x = points_x[maskz]
    points_y = points_y[maskz]
    points_z = points_z[maskz]
    points_z = points_z - points_z.min()
    loc_points = np.stack((points_x, points_y, points_z), axis=1)
    
    # find the points with a distance of 4 um to the center
    distance = np.sqrt(points_x**2 + points_y**2)
    mask = (distance>=dist) & (distance<dist+thickness)
    points4x = points_x[mask]
    points4y = points_y[mask]
    points4z = points_z[mask]
    
    theta = np.degrees(np.arctan2(points4y, points4x))
    length = 2*np.pi*dist*theta/360
    max_length = 20e3
    
    return length/1e3, points4z/1e3

def KD_panorama(dist, thickness):
    
    ### load point cloud data =================================================================================
    points = pipeline(fn_KD)
#     x0, y0 = 631, 772
#     x1, y1 = 189, 1334
#     ps = 5e4/1163
    xc = (x0kd-x1)*ps
    yc = (y1-y0kd)*ps
    points_x = points['x'] - xc
    points_y = points['y'] - yc
    points_z = points['z']
    mid_z = (points_z.min() + points_z.max())/2
    zmin = mid_z - 2e3
    zmax = mid_z + 2e3
    maskz = (points_z>=zmin) & (points_z<=zmax)
    points_x = points_x[maskz]
    points_y = points_y[maskz]
    points_z = points_z[maskz]
    points_z = points_z - points_z.min()
    loc_points = np.stack((points_x, points_y, points_z), axis=1)
    
    # find the points with a distance of 4 um to the center
    distance = np.sqrt(points_x**2 + points_y**2)
    mask = (distance>=dist) & (distance<dist+thickness)
    points4x = points_x[mask]
    points4y = points_y[mask]
    points4z = points_z[mask]
    
    theta = np.degrees(np.arctan2(points4y, points4x))
    length = 2*np.pi*dist*theta/360
    max_length = 20e3
    
    return length/1e3, points4z/1e3
```
